#### Tabla de valores para calcular qi

|   qi   |      EC(us/m)     | SAR ((mmol.L^-1)^0.5) | Na (meq/L)   | Cl (meq/L)    | HCO3 (meq/L)       |
|:------:|:-----------------:|:---------------------:|--------------|---------------|--------------------|
| 85-100 |     [200,750)     |         [2,3]         | [2,3)        | [1,4)         | [1,1.5)            |
| 60-85  |     [750,1500)    |         [3,6)         | [3,6)        | [4,7)         | [1.5,4.5)          |
| 35-60  | [1500,3000)       | [6,12)                | [6,9)        | [7,10)        | [4.5,8.5)          |
| 0-35   | EC<200 o EC>=3000 | SAR<2 o SAR>=12       | Na<2 o Na>=9 | Cl<1 o Cl>=10 | HCO3<1 o HCO3>=8.5 |

```python
qi_ranges = {
    "EC": [
        {"qi_min": 85.0, "qi_max": 100.0, "x_inf": 200.0, "x_sup": 750.0},
        {"qi_min": 60.0, "qi_max": 85.0, "x_inf": 750.0, "x_sup": 1500.0},
        {"qi_min": 35.0, "qi_max": 60.0, "x_inf": 1500.0, "x_sup": 3000.0},
        {"qi_min": 0.0, "qi_max": 35.0},
    ],
    "SAR": [
        {"qi_min": 85.0, "qi_max": 100.0, "x_inf": 2.0, "x_sup": 3.0},
        {"qi_min": 60.0, "qi_max": 85.0, "x_inf": 3.0, "x_sup": 6.0},
        {"qi_min": 35.0, "qi_max": 60.0, "x_inf": 6.0, "x_sup": 12.0},
        {"qi_min": 0.0, "qi_max": 35.0},
    ],
    "Na+": [
        {"qi_min": 85.0, "qi_max": 100.0, "x_inf": 2, "x_sup": 3},
        {"qi_min": 60.0, "qi_max": 85.0, "x_inf": 3, "x_sup": 6},
        {"qi_min": 35.0, "qi_max": 60.0, "x_inf": 6, "x_sup": 9},
        {"qi_min": 0.0, "qi_max": 35.0},
    ],
    "Cl-": [
        {"qi_min": 85.0, "qi_max": 100.0, "x_inf": 1, "x_sup": 4},
        {"qi_min": 60.0, "qi_max": 85.0, "x_inf": 4, "x_sup": 7},
        {"qi_min": 35.0, "qi_max": 60.0, "x_inf": 7, "x_sup": 10},
        {"qi_min": 0.0, "qi_max": 35.0},
    ],
    "HCO3-": [
        {"qi_min": 85.0, "qi_max": 100.0, "x_inf": 1, "x_sup": 1.5},
        {"qi_min": 60.0, "qi_max": 85.0, "x_inf": 1.5, "x_sup": 4.5},
        {"qi_min": 35.0, "qi_max": 60.0, "x_inf": 4.5, "x_sup": 8.5},
        {"qi_min": 0.0, "qi_max": 35.0},
    ],
}
```

#### Cálculo de qi

```python
def qi(xij,parameter):
  if parameter in qi_ranges.keys():
    i = 0
    flag_interval = False
    while ((i<(len(qi_ranges[parameter])-1)) and (not flag_interval)):
      x_inf = qi_ranges[parameter][i]['x_inf']
      x_sup = qi_ranges[parameter][i]['x_sup']
      if xij>=x_inf and xij<x_sup:
        qi_min = qi_ranges[parameter][i]['qi_min']
        qi_max = qi_ranges[parameter][i]['qi_max']
        qi_amp = qi_max - qi_min
        x_amp = x_sup - x_inf
        flag_interval = True

        qi = qi_max - ((xij-x_inf)*qi_amp/x_amp)
      else:
        i+=1

    if not flag_interval:
      x_inf = qi_ranges[parameter][0]['x_inf']
      x_sup = qi_ranges[parameter][2]['x_sup']

      if xij<x_inf:
        qi=qi_ranges[parameter][3]['qi_min']
      elif xij>=x_sup:
        qi=qi_ranges[parameter][3]['qi_max']

  return qi

def qiNa(xij):
  return qi(xij,'Na+')
```
